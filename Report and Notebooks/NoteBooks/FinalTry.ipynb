```python
import numpy as np

neg_probabilities = data['neg'].tolist()
nut_probabilities = data['nut'].tolist()
pos_probabilities = data['pos'].tolist()

# Calculate cumulative probabilities
cumulative_probs = np.cumsum(np.array([neg_probabilities, nut_probabilities, pos_probabilities]), axis=0)

# Calculate thresholds for each rating level
thresholds = []
for i in range(1, 6):  # Rating levels 1 to 5
    threshold = np.percentile(cumulative_probs[-1], i * 20)  # Calculate the threshold based on percentiles
    thresholds.append(threshold)

# Define mapping function
def map_rating(neg_prob, nut_prob, pos_prob):
    cumulative_prob = neg_prob + nut_prob + pos_prob
    for i, threshold in enumerate(thresholds):
        if cumulative_prob <= threshold:
            return i + 1  # Rating levels start from 1

# Example usage
ratings = [map_rating(neg, nut, pos) for neg, nut, pos in zip(neg_probabilities, nut_probabilities, pos_probabilities)]
```

```python
X = newTest['compound']
Y = newTest['Rating']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
lr_model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_pred)
print("Logistic Regression Accuracy:", lr_accuracy)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_pred)
print("Decision Tree Accuracy:", dt_accuracy)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", rf_accuracy)
```
